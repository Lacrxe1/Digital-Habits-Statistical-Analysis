```python
# Based on our heatmap, we can find the independent variables that influence Phone Unlocks the most. 
# These are: Device Hours Per Day, Anxiety score, Depression score, Stress level, High risk flag, and Digital Dependence.

import statsmodels.api as sm

X = df[['device_hours_per_day', 'anxiety_score', 'depression_score', 'stress_level', 'high_risk_flag', 'digital_dependence_score']]
Y = df['phone_unlocks']

X = sm.add_constant(X)

model = sm.OLS(Y, X)
results = model.fit()
print(results.summary())
```

The equation for the **Multi-Variable Regression Model** is: $$y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_p X_p + \varepsilon$$

Where: 

$\beta_0$ is the constant

$X_1$ is the first independent variable;

$X_2$ is the second independent variable;

$X_p$ is the number of independent variables after $X_2$ ;

$\epsilon$ is the error term.

For my equation, we have 6 independent variables. The equation is: $$\hat{y} = 3.32 + \beta_1 \times 8.82 + \beta_2 \times 0.26 + \beta_3 \times 1 + \beta_4 \times{-0.72} + \beta_5 \times 1.69 + \beta_6 \times 1.96$$

Where:

$\hat{y}$ represents Phone Unlocks

$3.32$ is the constant;

$\beta_1$ is Device hours per day;

$\beta_2$ is Anxiety Score;

$\beta_3$ is Depression Socre;

$\beta_4$ is Stress level;

$\beta_5$ is High Risk Flag;

$\beta_6$ is Digital Dependence Score.

From the Regression output, we can see that the Digital Dependence and Device Hours by far has the highest influence of the number of times a person unlocks their phones every day (at a t-stat of 24 and 20 respectively). The p_value (below 0.05) of every variable except high risk flag and anxiety score confirm that the variables are statistically significant. Finally, the high R^2 of 0.76 tells us that 76% of the variation in Phone Unlocks can be explained, or attributed to the variations of the equation.

# **Training the Model using OLS** # 

We will implement a basic OLS ML technique using the equation we found using statsmodel which is: $$\hat{y} = 3.32 + \beta_1 \times 8.82 + \beta_2 \times 0.26 + \beta_3 \times 1 + \beta_4 \times{-0.72} + \beta_5 \times 1.69 + \beta_6 \times 1.96$$

Where: 

$\hat{y}$ represents Phone Unlocks

$3.32$ is the constant;

$\beta_1$ is Device hours per day;

$\beta_2$ is Anxiety Score;

$\beta_3$ is Depression Socre;

$\beta_4$ is Stress level;

$\beta_5$ is High Risk Flag;

$\beta_6$ is Digital Dependence Score.

```python
# CREATING THE MODEL USING OLS
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['device_hours_per_day', 'anxiety_score', 'depression_score', 'stress_level', 'high_risk_flag', 'digital_dependence_score']]
y = df['phone_unlocks']

#Splitting Data 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
#Training it

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
#Visualizing It
from sklearn.metrics import r2_score

plt.figure(figsize=(12, 8))

sns.scatterplot(x = y_test, y = y_test_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)


r2 = r2_score(y_test, y_test_pred)
plt.text(0.86, 0.95, f'R^2 = {r2:.3f}', transform=plt.gca().transAxes)

plt.xlabel('Actual Phone Unlocks')
plt.ylabel('Predicted Phone Unlocks')
plt.title("Actual vs Predicted Values")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Seeing the Residuals

residuals = y_test - y_test_pred

plt.figure(figsize=(11, 7))
sns.residplot(x=y_test_pred, y=residuals, lowess=True, scatter_kws={'alpha': 0.6})


plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.tight_layout()
plt.show()
print("All Done!")
```
